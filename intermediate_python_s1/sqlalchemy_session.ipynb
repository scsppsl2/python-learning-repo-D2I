```python
header = Table('header', metadata_903, autoload_with = engine_903)

with engine_903.connect() as con:
    stmt = select(header).where(or_(header.columns.ETHNIC == "REFU",
                                    header.columns.ETHNIC == "NOBT"))
   # print(stmt)
    results = con.execute(stmt).fetchmany(size=10)

print(results) #more complicated select (where & all)
```

```python
#select all of the book table where the book_id is either 1 or 2

book = Table('book', metadata_gravity, autoload_with = engine_gravity)

with engine_gravity.connect() as con:
    stmt = select(book).where(or_(book.columns.book_id == 1,
                                  book.columns.book_id == 2))
    
    results = con.execute(stmt).fetchall()

print(results)
```
